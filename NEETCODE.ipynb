```python
class Solution(object):
    def containsDuplicate(self, nums):
        dub = set()
        for i in nums:
            dub.add(i)
        if len(dub) == len(nums):
            return False
        else:
            return True
```

```python
from collections import Counter
class Solution(object):
    def isAnagram(self, s, t):
        a = Counter(s)
        b = Counter(t)
        if a == b:
            return True
        else:
            return False
#takes log(n)
```

```python
class Solution(object):
    def twoSum(self, nums, target):
        for i in range(0,len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i]+nums[j] == target:
                    return i,j #brute force n2
```

```python
class Solution(object):
    def twoSum(self, nums, target):
        num_map = {}  
        for i, num in enumerate(nums):  
            complement = target - num
            if complement in num_map:  
                return [num_map[complement], i]
            num_map[num] = i  #O(n)
```

```python
class Solution(object):
    def smallerNumbersThanCurrent(self, nums):
        result = []
        for i in range(len(nums)):
            temp = 0
            for j in range(len(nums)):
                if nums[i] > nums[j]:
                    temp += 1
            result.append(temp)
        return result #on2 bruteforce
```

```python
class Solution(object):
    def findDifference(self, nums1, nums2):
        a = []
        b = []
        numset1 = set(nums1)
        numset2 = set(nums2)
        for i in nums1:
            if i not in numset2 and i not in a:
                a.append(i)
        for j in nums2:
            if j not in numset1 and j not in b:
                b.append(j)
        return a,b #optimal soln O(m+n)
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stk = []
        mapper = {")" : "(", "}" : "{", "]" : "["}
        for i in s:
            if i not in mapper:
                stk.append(i)
            else:
                if stk and stk[-1] == mapper[i]:
                        stk.pop()
                else:
                    return False
        return True if not stk else False
```

```python
class MinStack:
    def __init__(self):
        self.stk = []

    def push(self, val: int) -> None:
        self.stk.append(val)

    def pop(self) -> None:
        if self.stk:
            self.stk.pop(-1)

    def top(self) -> int:
        if self.stk:
            return self.stk[-1]
        else:
            None
    def getMin(self) -> int:
        if self.stk:
            return min(self.stk)
        else:
            None
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
def printList(head):
    curr = head
    while curr != None:
        print(curr.data, end = " -> ")
        curr = curr.next

#driver code
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
ans = printList(head)
```

## SEARCHING in LINK LIST


```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
def printer(head , x):
    curr = head
    count = 1
    while curr != None:
        if curr.data == x:
            return count
        curr = curr.next
        count += 1
    return -1
        
#driver code
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
x = 22
ans = printer(head , x)
print(ans)
```

## ADDING IN THE BEGINNING

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
def insertbegin(head, key):
    temp = Node(key)
    temp.next = head
    return temp
head = None
head = insertbegin(head, 10)
head = insertbegin(head, 20)
head = insertbegin(head, 30)
head = insertbegin(head, 40)
print(head.data)
```

# ADDING IN THE LAST

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None       
def insertend(head, key):
    if head == None:
        return Node(key)
    curr = head
    while curr.next != None:
        curr = curr.next
    curr.next = Node(key)
    return head
head = None
head = insertend(head, 10)
head = insertend(head, 20)
head = insertend(head, 30)
head = insertend(head, 40)
printList(head)
```

## SEARCHING


```python
def bs(arr, number):
    low = 0
    high = len(arr) - 1
    while(low <= high):
        mid = (low + high) // 2
        if arr[mid] == number:
            return mid
        elif arr[mid] > number:
            high = mid - 1
        else:
            low = mid + 1
    return None
arr = [1,3,5,7,9]
ans = bs(arr, 9)
print(ans)
```
